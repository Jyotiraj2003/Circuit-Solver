## Loop Method

```python
def factorial(N):
    if N > 0:
        fact = 1
        while N > 0:
            fact = fact * N
            N = N -1
        return fact
    
num = int(input("Enter a number:"))
print(factorial(num))

%timeit factorial(num)
```

Here we calculate the factorial of a number N, which is taken as input using loop method. We are taking a variable 'fact' as 1 if N > 0, as we can calculate factorial for positive number only.
Then we multiplying 'fact' with N and decreasing N by 1 and repreating the same process till N becomes 0.
At last we are calulating it's time using magical syntax of 'timeit' library.

## Recursion Method

```python
def factorial2(N):
    if N >= 0:
        if N == 0:
            return 1
        else:
            return factorial2(N-1) * N

num2 = int(input("Enter a number:"))
print(factorial2(num2))

%timeit factorial2(num2)
    
```

Here we are using recursion method for calculating the factorial of a number N which is taken as input. 
When we have N >= 0 we are proceeding in the function because factorial for non_zero number can only be calculated. Inside the function of N we are calling function of N-1 and multiplying it with N to find the factorial. 
For breaking the loop we are using ' if N ==0 , return 1'.
At last we are using magical syntax of 'timeit' to calculate the time.

```python
import numpy as np
def matrix_solver(a,b,r,c):
    
    for i in range(r):
        if a[i][i] == 0:
            if i == r-1:
                break
            a[i],a[i+1] = a[i+1],a[i]
        norm = a[i][i]
        for k in range(c):
            a[i][k] = a[i][k]/norm
        b[i] = b[i]/norm
        for j in range(r):
            if j != i:
                deno = a[j][i]
                for k in range(len(a[j])):
                    a[j][k] = a[j][k] - (a[i][k] * deno)
                b[j] = b[j] - (b[i] * deno)
        
    if a[r-2][r-2] == 0:
        if b[r-1] == 0:
            print("Infinite solutions are possible")
        else:
            print("No solution is possible")

    print(a)
    print(b)
    
    
try:
    row = int(input("Enter the number of rows:"))
    column =int(input("Enter the number of columns:"))
except Exception:
     print("The input given is not an integr. Please check and input an integer")


A = np.random.rand(row,column)
B = np.random.rand(row)
        
matrix_solver(A,B,row,column)
```

Inside the function we are first checking the digonal elements of matrix A. If it is 0 we are swapig the row with another row with non_zero element. We are swaping the corresponding rows of matricx B also.
Then we are normalizing the digonal element to 1 by dividing the whole row with the diagonal element. Here we are dividing the matrix B with the diagonla element.
Then we are making the other elements of the column of the normalized diagonal element by subtracting the normalized row from the required row  by multyiplying normalized row with corresponding column element.We are performing the exact operation in matrix B also.
We have put the above steps in loop which will run the no of rows times.

At last to check infinite solutions and no solution, we judge the second last diagonal element. If it is 0 and the last element of B is 0 then infinite solutions. If it is 0 and the last element of B is not 0 then no solutions.

```python
import numpy as np
def circuit_solver(ckt_name):
    with open(ckt_name,'r') as f:
        lines = f.read()
    r_n = []
    v_n = []
    i_n = []
    R = []
    V = []
    C = []
    count = 0

    cric = lines.splitlines()
    
    #frequency detection
    for l in cric:
        m = l.split()

        if m[0] == '.ac':
            fq = float(m[2])
            
    for l in cric:
        m = l.split()
        #for resistor 
        if l[0] == 'R':
            R.append(float(m[3]))
            for i in range(1,len(m)):
                if m[i] == 'GND':
                    m[i] = 0
                elif m[i] == 'n1':
                    m[i] = 1
                elif m[i] == 'n2':
                    m[i] = 2   
                elif m[i] == 'n3':
                    m[i] = 3       
            m[1] = int(m[1])
            m[2] = int(m[2])
            r_n.append([m[1],m[2]])
            
        #for voltage source
        if l[0] == 'V':
            if m[3] == 'dc': #dc voltage
                V.append(float(m[4]))
                count = count + 1

            if m[3] == 'ac': #ac voltage
                V.append(complex(float(m[4]) * np.cos(float(m[5])), float(m[4]) * np.sin(float(m[5]))))
                count =    count + 1

            for i in range(1,len(m)):
                if m[i] == 'GND':
                    m[i] = 0
                elif m[i] == 'n1':
                    m[i] = 1
                elif m[i] == 'n2':
                    m[i] = 2   
                elif m[i] == 'n3':
                    m[i] = 3 
            m[1] = int(m[1])
            m[2] = int(m[2])
            v_n.append([m[1],m[2]])

        #for current source
        if l[0] == 'I':
            if m[3] == 'dc': #dc current
                C.append(float(m[4]))

            if m[3] == 'ac': #ac current
                C.append(complex(float(m[4]) * np.cos(float(m[5])), float(m[4]) * np.sin(float(m[5]))))
            
            for i in range(1,len(m)):
                if m[i] == 'GND':
                    m[i] = 0
                elif m[i] == 'n1':
                    m[i] = 1
                elif m[i] == 'n2':
                    m[i] = 2   
                elif m[i] == 'n3':
                    m[i] = 3 
            m[1] = int(m[1])
            m[2] = int(m[2])   
            i_n.append([m[1],m[2]])
            
        #for inductor
        if l[0] == 'L':
            R.append(complex(0, 2*np.pi*fq*float(m[3])))
            for i in range(1,len(m)):
                if m[i] == 'GND':
                    m[i] = 0
                elif m[i] == 'n1':
                    m[i] = 1
                elif m[i] == 'n2':
                    m[i] = 2   
                elif m[i] == 'n3':
                    m[i] = 3       
            m[1] = int(m[1])
            m[2] = int(m[2])
            r_n.append([m[1],m[2]])

        #for capacitor   
        if l[0] == 'C':
            R.append(1/(complex(0, 2*np.pi*fq*float(m[3]))))
            for i in range(1,len(m)):
                if m[i] == 'GND':
                    m[i] = 0
                elif m[i] == 'n1':
                    m[i] = 1
                elif m[i] == 'n2':
                    m[i] = 2   
                elif m[i] == 'n3':
                    m[i] = 3       
            m[1] = int(m[1])
            m[2] = int(m[2])
            r_n.append([m[1],m[2]])


   #finding no of nodes
    for i in range(len(r_n)):
        node_no = max(r_n[i][0],r_n[i][1])
    for i in range(len(v_n)): 
         node_no = max(node_no,v_n[i][0],v_n[i][1])
    for i in range(len(i_n)): 
         node_no = max(node_no,i_n[i][0],i_n[i][1])


   #matrices intializing
    A = [[0 for cols in range(node_no + count)] for rows in range(node_no + count)]
    B = [0 for rows in range(node_no + count)]  

    #putting matrix values if resistor,inductor,capacitor is connected
    for i in range(len(r_n)):
        for j in range(2):
            if r_n[i][0] != 0 and r_n[i][1] != 0:
                A[(r_n[i][j]-1)][(r_n[i][j]-1)] += 1/(R[i])
                A[(r_n[i][j]-1)][(r_n[i][1-j]-1)] -= 1/(R[i])
            elif r_n[i][j] == 0:
                A[(r_n[i][1-j]-1)][(r_n[i][1-j]-1)] += 1/(R[i])

     #putting matrix values if voltage source is connected
    for i in range(len(v_n)):
        for j in range(2):
            if v_n[i][j] == 0:
                A[(v_n[i][1-j]-1)][node_no + i] +=1
                A[node_no + i][(v_n[i][1-j]-1)] +=1
        if v_n[i][0] != 0 and v_n[i][1] != 0:
            A[(v_n[i][0]-1)][node_no + i] +=1
            A[node_no + i][(v_n[i][0]-1)] +=1

            A[(v_n[i][1]-1)][node_no + i] -=1
            A[node_no + i][(v_n[i][1]-1)] -=1

        B[node_no + i] += V[i]

     #putting matrix values if current source is connected
    for i in range(len(i_n)):
        if i_n[i][0] == 0:
            B[(i_n[i][1]-1)] += C[i]
        if i_n[i][1] == 0:
            B[(i_n[i][0]-1)] -= C[i]
        if i_n[i][0] != 0 and i_n[i][1] !=0:
            B[(i_n[i][0]-1)] -= C[i]
            B[(i_n[i][1]-1)] += C[i]

    
    print(A)
    print(B)  
    matrix_solver(A,B,(node_no+count),(node_no+count))

circuit_solver("ckt1.netlist")
```
